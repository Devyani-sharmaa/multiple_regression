```python
import numpy as np
```

```python
# genrate dataset

x = 6 * np.random.rand(200, 1) - 3
y = 0.8 * x**2 + 0.9 * x + 2 + np.random.randn(200, 1)
```

```python
x
```

```python
y
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(x, y)
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2)
```

```python
x.shape
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
plt.scatter(x_train, y_train, color = "red", label = "Traning Data")
plt.scatter(x_test, y_test, color = "green", label = "Testing Data")
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.legend()
```

```python
# simple linear regression    
```

```python
from sklearn.linear_model import LinearRegression
```

```python

linear_model = LinearRegression()
```

```python
linear_model.fit(x_train , y_train)
```

```python
plt.scatter(x_train, y_train, color = "red", label = "Traning Data")
plt.scatter(x_test, y_test, color = "green", label = "Testing Data")
plt.plot(x_test, linear_model.predict(x_test), color = "blue", label = "predicted line")
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.legend()
```

```python
y_pred = linear_model.predict(x_test)
```

```python
from sklearn.metrics import r2_score
```

```python
r2_score(y_test, y_pred)
```

```python
plt.scatter(x_train, y_train, color = "red", label = "Traning Data")
plt.scatter(x_test, y_test, color = "green", label = "Testing Data")
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.legend()
```

```python
# ploynomial feature
```

```python
x
```

```python
from sklearn.preprocessing import PolynomialFeatures
```

```python
poly = PolynomialFeatures(degree = 10)
```

```python
x_train_new = poly.fit_transform(x_train)
```

```python
x_test_new = poly.fit_transform(x_test)
```

```python
x_test
```

```python
x_test_new
```

```python
ploy_model = LinearRegression()
```

```python
ploy_model.fit(x_train_new, y_train)
```

```python
y_pred_poly = ploy_model.predict(x_test_new)
```

```python

r2_score(y_pred_poly, y_test)
```

```python
plt.scatter(x_train, y_train, color = "red", label = "Traning Data")
plt.scatter(x_test, y_test, color = "green", label = "Testing Data")
plt.plot(x_test, y_pred_poly, color = "blue", label = "Predicted Line")
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.legend()
```

```python
x_new = np.linspace(-3, 3, 200).reshape(200, 1)
x_new_ploy = poly.fit_transform(x_new)
y_new = ploy_model.predict(x_new_ploy)
```

```python
plt.scatter(x_train, y_train, color = "red", label = "Traning Data")
plt.scatter(x_test, y_test, color = "green", label = "Testing Data")
plt.plot(x_new, y_new, color = "blue", label = "Predicted Line", linewidth = 3)
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.legend()
```
